```python
'''Handling missing data is crucial in the feature engineering process for the wine quality dataset. Here are common techniques and their pros and cons:

### **1. Mean/Median Imputation**
   - **Description:** Replace missing values with the mean (for continuous data) or median of the non-missing values in the feature.
   - **Advantages:**
     - Simple to implement.
     - Maintains the dataset's size.
   - **Disadvantages:**
     - Can distort the distribution of the data.
     - Does not capture variability or relationships between features.

### **2. Mode Imputation**
   - **Description:** Replace missing values with the most frequent value (mode) for categorical data.
   - **Advantages:**
     - Simple and effective for categorical variables.
   - **Disadvantages:**
     - Can introduce bias if the mode is not representative of the missing values.

### **3. Predictive Imputation**
   - **Description:** Use machine learning models (e.g., regression, k-NN) to predict and impute missing values based on other features.
   - **Advantages:**
     - Accounts for relationships between features, potentially leading to more accurate imputations.
   - **Disadvantages:**
     - Computationally intensive.
     - Model accuracy depends on the quality and quantity of available data.

### **4. K-Nearest Neighbors (k-NN) Imputation**
   - **Description:** Impute missing values using the values from the k-nearest neighbors (based on distance metrics) in the dataset.
   - **Advantages:**
     - Considers the similarity between data points, leading to potentially more accurate imputations.
   - **Disadvantages:**
     - Computationally expensive, especially for large datasets.
     - The choice of k and distance metric can impact results.

### **5. Interpolation**
   - **Description:** Estimate missing values by interpolating between known values, often used in time series data.
   - **Advantages:**
     - Can provide a smooth and reasonable estimate based on trends.
   - **Disadvantages:**
     - May not be suitable for all types of data, particularly if data points are sparse.

### **6. Deletion**
   - **Description:** Remove rows or columns with missing values.
   - **Advantages:**
     - Simplifies analysis and avoids potential biases from imputation.
   - **Disadvantages:**
     - Can lead to loss of valuable data, especially if many values are missing.

**Handling Missing Data in the Wine Quality Dataset:**
- **For continuous features (e.g., Alcohol, pH):** Mean or median imputation might be used.
- **For categorical features (e.g., Weather Condition, Wind Direction):** Mode imputation is common.

**Choice of Technique:**
- **Mean/Median Imputation:** Quick and straightforward but can affect the distribution.
- **Predictive or k-NN Imputation:** More sophisticated, especially if relationships between features are complex, but requires more computation.

In practice, the choice of imputation method depends on the amount and type of missing data, the impact of missing values on the analysis, and the computational resources available.'''
```

```python
'''Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?'''
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('winequality-red.csv')

# Separate features and target variable
X = data.drop('quality', axis=1)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Cumulative explained variance
cumulative_explained_variance = explained_variance_ratio.cumsum()

# Plot cumulative explained variance
plt.figure(figsize=(8, 5))
plt.plot(cumulative_explained_variance, marker='o')
plt.axhline(0.90, color='r', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
# Find number of components that explain at least 90% variance
num_components = (cumulative_explained_variance >= 0.90).argmax() + 1
print(f'Minimum number of principal components required to explain 90% of the variance: {num_components}')

```
